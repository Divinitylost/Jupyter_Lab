```python
#Customisation

year = [1950,1970,1990,2010]
pop = [2.5,3.6,5.2,6.9]

plt.plot(year,pop) #first arguement is horizontal (x) second arguement (y)

#Titles
plt.xlabel("Years")
plt.ylabel("Population")
plt.title("World Population Projection")

#Axis Start Position
plt.yticks([0,2,4,6,8],["0","2B","4B","6B","8B"])
#the first arguement is the y axis ticks and then the name is changed with the second arg

#Alternative Method
#help(plt.xticks)
#ticks = [0,2,4,6,8]
#desc = ["0","2B","4B","6B","8B"]
#plt.yticks(ticks,desc)

plt.show()

#Append Date to Year and Pop

year = [1800,1850,1900]+year
pop = [1,1.2,1.6]+pop

plt.plot(year,pop)

#Titles
plt.xlabel("Years")
plt.ylabel("Population")
plt.title("World Population Projection")

#Axis Start Position
plt.yticks([0,2,4,6,8],["0","2B","4B","6B","8B"])

plt.show()
```

```python
# Definition of dictionary
europe = {'spain':'madrid', 'france':'paris', 'germany':'berlin', 'norway':'oslo' }

# Add italy to europe
europe["italy"]="rome"

# Print out italy in europe
print("italy" in europe)

# Add poland to europe
europe["poland"]="warsaw"

# Print europe
print(europe)
```

```python
# Dictionary of dictionaries
europe = { 'spain': { 'capital':'madrid', 'population':46.77 },
           'france': { 'capital':'paris', 'population':66.03 },
           'germany': { 'capital':'berlin', 'population':80.62 },
           'norway': { 'capital':'oslo', 'population':5.084 } }


# Print out the capital of France
print(europe["france"])

# Create sub-dictionary data
data = {"capital":"rome", "population":59.83}

# Add data to europe under key 'italy'
europe["italy"]=data

# Print europe
print(europe)
```
